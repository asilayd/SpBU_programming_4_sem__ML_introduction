```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import time
```

Мы используем датасет Digits из scikit-learn, который имеет 10 классов (цифры от 0 до 9) и 64 признака (размер 8x8 пикселей).

```python
digits = load_digits()
X = digits.data
y = digits.target

# Разделение на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
digits
```

```python
# Дерево решений
tree_clf = DecisionTreeClassifier(random_state=42, max_depth =10)
start_time = time.time()
tree_clf.fit(X_train, y_train)
tree_time = time.time() - start_time
tree_pred = tree_clf.predict(X_test)
tree_acc = accuracy_score(y_test, tree_pred)

# SVM
svm_clf = SVC(random_state=42)
start_time = time.time()
svm_clf.fit(X_train, y_train)
svm_time = time.time() - start_time
svm_pred = svm_clf.predict(X_test)
svm_acc = accuracy_score(y_test, svm_pred)

# KNN
knn_clf = KNeighborsClassifier()
start_time = time.time()
knn_clf.fit(X_train, y_train)
knn_time = time.time() - start_time
knn_pred = knn_clf.predict(X_test)
knn_acc = accuracy_score(y_test, knn_pred)
```

```python
#Сравнение точности и эффективности
print("Decision Tree:")
print(f"Accuracy: {tree_acc:.4f}")
print(f"Time taken: {tree_time:.4f} seconds")

print("\nSVM:")
print(f"Accuracy: {svm_acc:.4f}")
print(f"Time taken: {svm_time:.4f} seconds")

print("\nKNN:")
print(f"Accuracy: {knn_acc:.4f}")
print(f"Time taken: {knn_time:.4f} seconds")
```
