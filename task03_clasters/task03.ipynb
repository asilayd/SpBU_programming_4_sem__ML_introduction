```python
import numpy as np
from sklearn.datasets import make_blobs, make_moons, make_circles

# Генерация выборок разного размера
def generate_samples(n_samples):
    blobs = make_blobs(n_samples=n_samples, centers=3, cluster_std=0.60, random_state=0)
    moons = make_moons(n_samples=n_samples, noise=0.1, random_state=0)
    circles = make_circles(n_samples=n_samples, noise=0.1, factor=0.3, random_state=0)
    return blobs, moons, circles

# Примеры генерации выборок
blobs_100, moons_100, circles_100 = generate_samples(100)
blobs_500, moons_500, circles_500 = generate_samples(500)
blobs_1000, moons_1000, circles_1000 = generate_samples(1000)
```

Индекс силуэта (silhouette_score) измеряет, насколько хорошо объекты вписываются в свои кластеры. Значения варьируются от -1 до 1, где 1 означает отличное разделение, а -1 указывает на плохое разделение.

```python
from sklearn.cluster import KMeans, DBSCAN
from sklearn.cluster import SpectralClustering
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score

def cluster_samples(X, method):
    if method == 'kmeans':
        model = KMeans(n_clusters=3)
    elif method == 'spectral':
        model = SpectralClustering(n_clusters=2, affinity='nearest_neighbors')
    elif method == 'dbscan':
        model = DBSCAN(eps=0.3, min_samples=5)
    else:
        raise ValueError("Unsupported method")

    labels = model.fit_predict(X)
    return labels

def plot_clusters(X, labels, title):
    plt.figure(figsize=(12, 6))

    # Исходные данные
    plt.subplot(1, 2, 1)
    plt.scatter(X[:, 0], X[:, 1], c='blue', marker='o')
    plt.title('Original Data')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')

    # Результаты кластеризации
    plt.subplot(1, 2, 2)
    scatter = plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o')
    plt.title(f'Clustered Data')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')

    # Добавление цветовой шкалы
    plt.colorbar(scatter, ax=plt.gca())

    # Расчет и отображение индекса силуэта
    if len(set(labels)) > 1:
        score = silhouette_score(X, labels)
        plt.suptitle(f'{title}\nSilhouette Score: {score:.2f}')
    else:
        plt.suptitle(f'{title}\nSilhouette Score: Not enough clusters')

    plt.tight_layout()
    plt.show()

# Пример кластеризации и визуализации

for sample_size in [100, 500, 1000]:
    blobs, moons, circles = generate_samples(sample_size)

    for method in ['kmeans', 'spectral', 'dbscan']:
        for data, name in zip([blobs[0], moons[0], circles[0]], ['blobs', 'moons', 'circles']):
            labels = cluster_samples(data, method)
            plot_clusters(data, labels, f'{name.capitalize()} - {method.capitalize()} (n={sample_size})')
```

#### Графовый подход (SpectralClustering):

Подходит для данных, где кластеры имеют сложную форму или взаимосвязь между точками сложна.
Эффективен для небольших и средних наборов данных.

#### Центроидный подход (KMeans):

Подходит для данных, где кластеры имеют круглую или сферическую форму.
Требует заранее заданного количества кластеров.
Эффективен для больших наборов данных, но может не справиться с кластерами неправильной формы.

#### Оценка плотности (DBSCAN):

Подходит для данных с неравномерной плотностью и сложной формой.
Не требует заранее заданного количества кластеров.
Чувствителен к выбору параметров eps и min_samples.
