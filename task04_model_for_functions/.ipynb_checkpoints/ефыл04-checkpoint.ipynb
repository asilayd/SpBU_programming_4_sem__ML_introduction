## Нейросеть



Можно построить нейросеть для приближения целевой одномерной функции. Мы будем использовать PyTorch для создания регрессионной модели, где сеть будет обучаться на основе сгенерированных данных с добавлением шума.

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt

# Генерация данных
def generate_data(func, n_points=100, noise=0.1):
    x = np.linspace(-2, 2, n_points)
    y = func(x) + noise * np.random.randn(n_points)
    return x, y

# Определение функций
functions = {
    'exp(x)': np.exp,
    'cos(x)': np.cos,
    'log(x)': lambda x: np.log(np.abs(x) + 0.1),  # добавляем 0.1 для предотвращения логарифма от 0
    'x^2': lambda x: x**2,
    'x^3': lambda x: x**3
}

# Модель
class SimpleNN(nn.Module):
    def __init__(self):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(1, 64)
        self.fc2 = nn.Linear(64, 64)
        self.fc3 = nn.Linear(64, 1)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = self.fc3(x)
        return x

# Функция для обучения модели
def train_model(func_name, target_function, n_epochs=1000, noise=0.1):
    # Генерация данных
    x, y = generate_data(target_function, noise=noise)
    x_train = torch.tensor(x, dtype=torch.float32).view(-1, 1)
    y_train = torch.tensor(y, dtype=torch.float32).view(-1, 1)
    
    # Инициализация модели, функции потерь и оптимизатора
    model = SimpleNN()
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=0.01)

    # Обучение модели
    for epoch in range(n_epochs):
        model.train()
        optimizer.zero_grad()
        outputs = model(x_train)
        loss = criterion(outputs, y_train)
        loss.backward()
        optimizer.step()

    # Тестирование модели
    model.eval()
    with torch.no_grad():
        predictions = model(x_train).numpy()

    return x, y, predictions,target_function(x), func_name

# Тренируем модели на нескольких функциях
results = []
for func_name, target_function in functions.items():
    result = train_model(func_name, target_function)
    results.append(result)

# Визуализация
plt.figure(figsize=(15, 10))
for i, (x, y, predictions, true_function, func_name) in enumerate(results):
    plt.subplot(2, 3, i+1)
    plt.scatter(x, y, label='Noisy Data')
    plt.plot(x, predictions, color='red', label='Predicted Function')
    plt.title(f'Model for {func_name}')
    plt.legend()

plt.tight_layout()
plt.show()
```

## Без нейросети

Для подбора подходящей регрессионной модели к заданной одномерной функции можно использовать несколько стратегий, включая полиномиальную регрессию и нелинейные трансформации. Мы можем реализовать процесс следующим образом:

Сгенерируем данные для обучения, используя одну из выбранных функций (например, exp(x), cos(x), log(x), и т.д.).
Добавим шум к данным, чтобы смоделировать реальные условия.
Попробуем несколько моделей регрессии: линейную регрессию, полиномиальную регрессию и другие, соответствующие преобразованию данных.
Оценим качество моделей с помощью метрик (например, среднеквадратичной ошибки).

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error

# Генерация данных для функции
def generate_data(func, x_range, n_points=100, noise=0.1):
    X = np.linspace(x_range[0], x_range[1], n_points).reshape(-1, 1)
    y = func(X).flatten()
    # Замена значений NaN на 0 и добавление шума
    y = np.nan_to_num(y) + noise * np.random.randn(n_points)
    return X, y

# Линейная регрессия
def fit_linear(X, y):
    model = LinearRegression()
    model.fit(X, y)
    return model

# Полиномиальная регрессия
def fit_polynomial(X, y, degree=2):
    poly = PolynomialFeatures(degree)
    X_poly = poly.fit_transform(X)
    model = LinearRegression()
    model.fit(X_poly, y)
    return model, poly

# Оценка модели
def evaluate_model(model, X, y, poly=None):
    if poly is not None:
        X = poly.transform(X)
    y_pred = model.predict(X)
    mse = mean_squared_error(y, y_pred)
    return mse



# Основная программа для нескольких функций
def train_and_plot_for_functions(functions, function_names, x_range=(-2, 6), noise=0.1):
    for func, name in zip(functions, function_names):
        # Генерация данных для функции
        X, y = generate_data(func, x_range, noise=noise)
        
        # Линейная регрессия
        linear_model = fit_linear(X, y)
        linear_mse = evaluate_model(linear_model, X, y)
        
        # Полиномиальная регрессия (степень 3)
        poly_model, poly_transformer = fit_polynomial(X, y, degree=3)
        poly_mse = evaluate_model(poly_model, X, y, poly=poly_transformer)
        
        # Отображение результатов
        plt.scatter(X, y, label='Data', color='blue', s=7)
        plt.plot(X, linear_model.predict(X), label=f'Linear (MSE={linear_mse:.2f})', color='red')
        
        X_poly_pred = poly_transformer.transform(X)
        plt.plot(X, poly_model.predict(X_poly_pred), label=f'Polynomial (MSE={poly_mse:.2f})', color='green')
        
        plt.legend()
        plt.title(f"Regression for {name}")
        plt.show()

        print(f"Results for {name}:")
        print(f"  Linear MSE: {linear_mse}")
        print(f"  Polynomial MSE: {poly_mse}")
        print("-" * 40)

# Список функций для тестирования
functions = [np.exp, np.cos, lambda x: np.log(x+2.1), lambda x: x**2, lambda x: np.power(x, 3)]
function_names = ['exp(x)', 'cos(x)', 'log(x+2.1)', 'x^2', 'x^3']

# Запуск программы
train_and_plot_for_functions(functions, function_names)
```
